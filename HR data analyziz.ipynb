```python
import pandas as pd
df = pd.read_csv("HR_analytics.csv")
df
```

```python
df.drop_duplicates(subset=['EmpID'])
df
# remove all rows that contain null values
df = df.dropna(how="any")
df
```

```python
print(df['EmployeeCount'].value_counts())
print(df['Over18'].value_counts())
age_groups = df['AgeGroup'].value_counts()
age_groups
```

```python
df = df.drop(columns=["AgeGroup","EmployeeCount","Over18"])
df
```

```python
for column_name in ['Attrition','OverTime']:
    df[column_name] = df[column_name].replace({'Yes': 1, 'No': 0})
df
```

```python
df['BusinessTravel'].value_counts()
```

```python
df['BusinessTravel'] = df['BusinessTravel'].replace({'Travel_Frequently': 1,'Travel_Rarely':0.5,'TravelRarely':0.5, 'Non-Travel': 0})
df
```

```python
print(df['Department'].value_counts())
print(df['EducationField'].value_counts())
```

```python
print(df['SalarySlab'].value_counts())
```

```python
print(df['Gender'].value_counts(),"\n")
print(df['MaritalStatus'].value_counts(),"\n")
print(df['JobRole'].value_counts(),"\n")
print(df['StandardHours'].value_counts())
```

```python
standard_hours = 80
df = df.drop(columns="StandardHours")
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace({'Married': 1, 'Single': 0,'Divorced':0})
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0})
```
